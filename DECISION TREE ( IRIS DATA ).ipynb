# DECISION TREE CLASSIFIER FROM SKLEARN- LEARN ON IRIS DATASETS:

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

# 1. Load the data

```python
iris = datasets.load_iris()
```

```python
iris
```

# 2. Features and target are in diffrent

```python
x = iris.data
y = iris.target
x.ndim,y.ndim
```

# 3. Split the data into training and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

# 4. Standardize the Data

```python
sc = StandardScaler()
x_train_std = sc.fit_transform(x_train)
x_test_std = sc.fit_transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

# 5. Choose Algo & Set the hyperparametres

```python
tree = DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
tree
```

# 6. Train the model

```python
tree.fit(x_train_std,y_train)
```

# 7. Make the prediction using the model

```python
pred = tree.predict(x_test_std)
```

```python
pred
```

```python
y_test
```

```python
np.where(pred!=y_test)
```

```python
pred[8],y_test[8]
```

```python
print('ACCURACY :', accuracy_score(y_test,pred))
```
